```python
from mss.darwin import MSS as mss
import os
from time import time, sleep
from PIL import Image
from resizeimage import resizeimage

sleep(15)

total=0
i=0
# take screenshot 
t=time()
elapsed=time()-t
while elapsed <= 950:
    t0=time()
    with mss() as sct:
        for filename in sct.save(mon=1,output='tmp/screenshot.jpg'):
            pass

    # resize screenshot and move to input folder
    with open('tmp/screenshot.jpg', 'r+b') as f:
        with Image.open(f) as image:
            cover = resizeimage.resize_cover(image, [1400, 788])
            cover.save('input/img'+str(int(100*time()))+'.jpg', image.format)
    
    dt=time()-t0
    i=i+1
    total=total+dt
    avtime=total/i
    print(avtime)
    elapsed=time()-t
```
