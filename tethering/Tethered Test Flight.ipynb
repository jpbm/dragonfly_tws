```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df=pd.read_excel('testflight_01_13042017.xlsx')
```

```python
data=df.loc[:32]
data=data.drop(31)
print(data)
```

```python
plt.figure(figsize=(14,7))
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)    
ax.spines["right"].set_visible(False)
plt.plot(data['Time of Day'],data['Remaining Charge [mAh]'],Marker='.')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_ylabel('Remaining Charge [mAh]',fontsize=16)
ax.set_xlabel('Time of Day',fontsize=16)
ax.set_title('Battery Charge vs. Time',fontsize=20)

y=range(2700,3450)
x=len(y)*['19:57']
plt.plot(x,y,Color='black',LineStyle=':',alpha=0.6)
plt.text('20:00',3450,'Take off',fontsize=14)
plt.savefig('BatteryChargeVSTime.png')
```

```python
plt.figure(figsize=(14,7))
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)    
ax.spines["right"].set_visible(False)
plt.plot(data['Time of Day'],data['Temperature [C]'],Marker='.')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_ylabel('Battery Temperature [C]',fontsize=16)
ax.set_xlabel('Time of Day',fontsize=16)
ax.set_title('Battery Temperature vs. Time',fontsize=20)

y=range(21,26)
x=len(y)*['19:57']
plt.plot(x,y,Color='black',LineStyle=':',alpha=0.6)
plt.text('20:00',25,'Take off',fontsize=14)
plt.savefig('BatteryTempVSTime.png')
```
