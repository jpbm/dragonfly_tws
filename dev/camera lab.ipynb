```python
import os 
from itertools import cycle
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
class Camera2(object):
    def __init__(self):
        self.frames=cycle([open('output/'+f,'rb').read() for f in os.listdir('output') if '.jpg' in f])

    def get_frame(self):
        return self.frames.__next__()
```

```python
c=Camera2()
```
